```python
from seabird.cnv import fCNV
```

```python
profile = fCNV('dar84-01_011.cnv')
```

```python
profile.attrs # It will return the header, as a dictionary
```

```python
profile.keys() # It will list the available variables
```

```python
# plot 'wetCDOM' x 'PRES' y

from matplotlib import pyplot as plt

plt.plot(profile['wetCDOM'], profile['PRES'],'g')
# plt.plot(profile['TEMP'], profile['PRES'],'b')

plt.gca().invert_yaxis()
plt.xlabel('WET Labs CDOM [mg/m^3]')
plt.ylabel('pressure [dbar]')
plt.title(profile.attrs['filename'])
```
